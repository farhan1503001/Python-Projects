```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
response=urlopen('http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/').read()
#Now for seeing the datas for easily
soup=BeautifulSoup(response,'html.parser')
#soup.prettify()
```

```python
#For finding the blocks that we want to find
property_htm=soup.find_all('div',{'class':'propertyRow'})\
#Now find the numnber of extraction
size=len(property_htm)
print(size)
#Now we observe an example
property_htm[0]
```

```python
#Now we find the price of the property
#print(property_htm[0].find_all('h4',{'class':'propPrice'}))
all=property_htm[0].find_all('h4',{'class':'propPrice'})[0].text.replace('\n','').replace(' ','')
print(all)
beds=property_htm[1].find_all('span',{'class':'infoBed'})[0].find('b').text
print(beds)
baths=property_htm[1].find_all('span',{'class':'infoValueFullBath'})[0].find('b').text
print(baths)
#Now finding the address
st_address=property_htm[1].find_all('span',{'class':'propAddressCollapse'})[0].text
m_address=property_htm[1].find_all('span',{'class':'propAddressCollapse'})[1].text
print(st_address)
print(m_address)
#print(type(all))
```

```python
feature_extractor=property_htm[1].find_all('div',{'class':'columnGroup'})
print(feature_extractor)
feature_extractor=feature_extractor[1].find_all('span',{'class':'featureGroup'})
print(' ')
print(feature_extractor)
```

```python
#Now for all entries of the first page we are printing and seeing if the data is working
for property in property_htm:
    print(property.find_all('span',{'class':'propAddressCollapse'})[0].text)
    print(property.find_all('span',{'class':'propAddressCollapse'})[1].text)
    try:
        print(property.find_all('h4',{'class':'propPrice'})[0].text.replace('\n','').replace(' ',''))
    except:
        print(None)
    try:
        print(property.find_all('span',{'class':'infoBed'})[0].find('b').text)
    except:
        print(None)
    try:
        print(property.find_all('span',{'class':'infoValueFullBath'})[0].find('b').text)
    except:
        print(None)
    for sample in property.find_all('div',{'class':'columnGroup'}):
        for group,value in zip(sample.find_all('span',{'class':'featureGroup'}),sample.find_all('span',{'class':'featureName'})):
            print(group.text,value.text)
    print('\n')
```

```python
l=[]
for property in property_htm:
    d={}
    d['Address']=property.find_all('span',{'class':'propAddressCollapse'})[0].text +','+property.find_all('span',{'class':'propAddressCollapse'})[1].text
    try:
        d['Price']=property.find_all('h4',{'class':'propPrice'})[0].text.replace('\n','').replace(' ','')
    except:
        d['Price']=None
    try:
        d['Area']=property.find_all('span',{'class':'infoSqFt'})[0].find('b').text
    except:
        d['Area']=None
    try:
        d['Beds']=property.find_all('span',{'class':'infoBed'})[0].find('b').text
    except:
        d['Beds']=None
    try:
        d['Full Baths']=property.find_all('span',{'class':'infoValueFullBath'})[0].find('b').text

    except:
        d['Full Baths']=None
    try:
        d['Half Baths']=property.find_all('span',{'class':'infoValueHalfBath'})[0].find('b').text

    except:
        d['Half Baths']=None
    for sample in property.find_all('div',{'class':'columnGroup'}):
        for group,value in zip(sample.find_all('span',{'class':'featureGroup'}),sample.find_all('span',{'class':'featureName'})):
            if 'Lot Size' in group.text:
                d['Lot size']=value.text
    l.append(d)
print(l)
    
```

```python
for item in l:
    print(item)
    print(' ')
```

```python
import pandas as pd
dataframe=pd.DataFrame(l)
```
